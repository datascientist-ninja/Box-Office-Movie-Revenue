```python
#Importing necessary library 

import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
regr = LinearRegression()
```

```python
#Importing of data using Pandas
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data.describe() #Summary statistics
```

```python
budget = data['production_budget_usd']
type(budget)
```

```python
#Transforming Series to Dataframe
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
#Regression Analysis
regr = LinearRegression()
regr.fit(X, y)
predict = regr.predict(X)
r_square = regr.score(X,y)
```

```python
%matplotlib inline 

plt.title('Film Cost v Global Revenue', fontsize=16)
plt.xlabel('Production Budget ($)', fontsize=14)
plt.ylabel('Worldwide Gross ($)', fontsize=14)


plt.scatter(X, y)
plt.show()
```
